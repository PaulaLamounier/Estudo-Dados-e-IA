```python
import pandas as pd
```

```python
idades = pd.Series([10, 50, 30, 80, 10], index=['a', 'b', 'c', 'd', 'e'], name='Idades')
print(idades)
```

```python
print (f"O total de registros de idade é: {idades.count()}")
print (f"A soma das idades é: {idades.sum()}")
print (f"A média das idades é: {idades.mean()}")
print (f"O valor mínimo das idades é: {idades.min()}")
print (f"O valor máximo das idades é: {idades.max()}")
print (f"O desvio padrão das idades é: {idades.std()}")
print (f"A variância das idades é: {idades.var()}")
print (f"O valor da mediana das idades é: {idades.median()}")
print (f"Os valores das idades ordenados são: {idades.sort_values().values}")
print (f"Os valores das idades ordenados pelo índice são: {idades.sort_index().values}")
print (f"Os 3 maiores valores das idades são: {idades.nlargest(3).values}")
print (f"Os 3 menores valores das idades são: {idades.nsmallest(3).values}")
print (f"A descrição estatística das idades é:\n{idades.describe()}")
print (f"A soma cumulativa das idades é: {idades.cumsum().values}")
print (f"O produto cumulativo das idades é: {idades.cumprod().values}")
print (f"As idades únicas são: {idades.unique()}")
print (f"A contagem de valores únicos das idades é:\n{idades.value_counts()}")
```

```python
dados = {
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'idade': [23, 35, 45, 29, 40],
    'salario': [5000, 7000, 8000, 6000, 9000]
}
df = pd.DataFrame(dados)
df["somaIdadeSalário"] = df["idade"] + df["salario"]
df["genero"] = ['F', 'M', 'M', 'F', 'M']
salario_por_genero = df.groupby("genero")["salario"].mean()
print(f"Média salarial por gênero:\n{salario_por_genero}")
salario_ordenado = df.sort_values(by="salario", ascending=False)
print(f"DataFrame ordenado por salário (decrescente):\n{salario_ordenado}")
df["salario"] = df["salario"].astype(float)

print(df)
```

```python
df2 = df.copy()
df2['salario'] = df2['salario'] * 1.2
print(df2)
```

```python
df3 = df2.copy()
df3['idade'] = df3['idade'] + 5
df3.loc[df3['genero'] == 'F', 'salario'] *= 1.9
df3.loc[(df3['genero'] == 'M') & (df3['salario'] <= 10000), 'salario'] *= 1.5
print(df3)
```
