```python
con = sqlite3.connect('teste.db')
cur = con.cursor()
```

```python
cur.execute('''
            CREATE TABLE IF NOT EXISTS pessoas (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                idade INTEGER NOT NULL,
                cidade TEXT NOT NULL,
                salário REAL NOT NULL)
            ''')
```

```python
cur.execute('''
INSERT INTO pessoas (nome, idade, cidade, salário) VALUES 
            ('Antônio', '25', 'Caldas Novas','2000'),
            ('Ana', 30, 'Salvador', 3800),
            ('Paula', 30, 'Rio de Janeiro', 3900)
             ''')           
```

```python
con.commit()
```

```python
cur.execute('''
SELECT * FROM pessoas where salário >= 3500 ORDER BY salário DESC
            ''')
resultado = cur.fetchall()
for linha in resultado:
    print(linha)
```

```python
cur.execute(''' delete from pessoas where id = 2 ''')
con.commit()
```
