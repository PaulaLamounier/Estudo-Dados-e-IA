```python
import sqlite3
```

```python
con = sqlite3.connect('teste.db')
cur = con.cursor()
```

```python
cur.execute('''
            CREATE TABLE IF NOT EXISTS categorias (
                id INTEGER PRIMARY KEY,
                nome TEXT NOT NULL)
            ''')
```

```python
cur.execute('''PRAGMA table_info(categorias)''')
estrutura = cur.fetchall()
for coluna in estrutura:
    print(coluna)
```

```python
cur.execute('''select name from sqlite_master where type = 'table' ''')
for table in cur.fetchall():
    print(table [0])
```

```python
cur.execute('''ALTER TABLE categorias ADD COLUMN ativo BOOLEAN DEFAULT 1''')
```

```python
cur.execute('''DROP TABLE categorias''')
```
