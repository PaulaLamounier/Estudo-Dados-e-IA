```python
pessoas = []
pessoas.append({
    'nome': 'João',
    'idade': 30,
    'cidade': 'São Paulo',
    'Assinante': 'True'
})
pessoas.append({
    'nome': 'Maria',
    'idade': 30,
    'cidade': 'Recife',
    'Assinante': 'False'
})
print(pessoas)
```

```python
for pessoa in pessoas:
    print(f"""Nome: {pessoa['nome']}, 
          Idade: {pessoa['idade']}, 
          Cidade: {pessoa['cidade']}""")
    if pessoa['Assinante'] == 'True':
        print(f"{pessoa['nome']} é assinante.")
        print('--------')
    else:
        print(f"{pessoa['nome']} não é assinante.")
# Exemplo de script Python com várias funcionalidades
```
