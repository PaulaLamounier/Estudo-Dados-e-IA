Os exercícios abaixo envolvem o uso da bliblioteca Math do Python. Todos foram extraídos do livro "Dominando Pyton" de Ohan Conceição.

```python
import math
```

```python
"""Algumas funções da biblioteca math em Python."""
print("Valor de pi:", math.pi)
print("Valor de e:", math.e)
print("Raiz quadrada de 16:", math.sqrt(16))
print("Fatorial de 5:", math.factorial(5))
print("Seno de 90 graus:", math.sin(math.radians(90)))
print("Coseno de 0 graus:", math.cos(math.radians(0)))
print("Tangente de 45 graus:", math.tan(math.radians(45)))
print("Logaritmo base 10 de 1000:", math.log10(1000))
print("Logaritmo natural de e:", math.log(math.e))
print("Arredondamento para cima de 4.3:", math.ceil(4.3))
print("Arredondamento para baixo de 4.7:", math.floor(4.7))
```

```python
'''Calcular e exibir o prrímetro de um círculo solicitando o raio ao usuário'''
raio = float(input("Digite o raio do círculo: "))
perim = 2 * math.pi * raio 
print("O perímetro do círculo é:", perim)
```

```python
'''Calcular a área do círculo a partir do raio utilizando a fórmula A = πr²'''
raio = float(input("Digite o raio do círculo: "))
area = math.pi * (raio ** 2)
print("A área do círculo é:", area)
```

```python
'''Calcular o delta de uma equação do segundo grau (ax² + bx + c = 0)'''
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c:"))
delta = b**2 - 4*a*c
print("O valor do delta é:", delta)
```
